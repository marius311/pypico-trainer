```python
import scipy.special
from pandas import *
import cosmoslik as K
import cosmoslik.plugins.derivers.hubble_theta.hubble_theta as H
```

```python
import os, cPickle
def cache(name,f=None,reset=False):
    try:
        if reset: raise Exception()
        with open('.cache/%s'%name) as f: return cPickle.load(f)
    except: 
        assert f!=None, "Value not cached, must provide function."
        r = f()
        if not os.path.exists('.cache'): os.makedirs('.cache')
        with open('.cache/%s'%name,'w') as f: cPickle.dump(r,f,protocol=2)
        return r
```

```python
chain=K.chains.load_chain('lowz.chain')
```

```python
ps=['tau', 'theta', 'omnuh2', 'ombh2', 'omch2', 'logA', 'ns', 'omk','w']
```

```python
covmat = chain.cov(ps)
```

```python
def fac(d):
    """ Ratio of the volume of unit hypercube to unit hypersphere in d dimensions""" 
    return 2**d/(2*pi**(d/2.)/scipy.special.gamma(d/2.)/d)
```

```python
def getwhite(dim,samp,sigma):
    return array(filter(lambda x: norm(x)<sigma, np.random.uniform(-sigma,sigma,(samp*fac(dim),dim))))[:samp]
```

```python
w_cps = cache('w_cps_80000_6',lambda: getwhite(dim=9,samp=80000,sigma=6))
```

```python
cps = DataMatrix(dot(w_cps,cholesky(covmat).T),columns=ps)+chain.mean(ps)
```

```python
cps=cps[cps['omnuh2']>0]
cps=cps[cps['w']<0]
cps=cps[cps['w']>-2]
```

```python
K.chains.likegrid([K.chains.Chain({k:cps[k] for k in ps}),chain],params=ps);
```

```python
def camb_derived(p):
    h = H.theta2hubble(p["theta"],p["ombh2"],p["omch2"],p["omk"],p['omnuh2'],p["w"],0,3.046)/100.
    return {"hubble":100*h,
            "ombh2":p["ombh2"],
            "omch2":p["omch2"],
            "omnuh2":p["omnuh2"],
            "omk":p["omk"],
            "scalar_amp(1)":exp(p["logA"])*10**(-10),
            "scalar_spectral_index(1)":p["ns"],
            "re_optical_depth":p["tau"],
            "theta":p["theta"],
            "logA":p["logA"],
            "w":p["w"]}
```

```python
from multiprocessing import Pool
```

```python
pool = Pool(8)
```

```python
d_cps = cache('d_cps_80000_6',lambda: DataMatrix(pool.map(camb_derived,[c for _,c in cps.iterrows()])))
```

```python
def savecps(file,names,dat):
    with open(file,'w') as f:
        f.write('#'+' '.join(names)+'\n')
        savetxt(f,[[d[n] for n in names] for _,d in dat.iterrows()])
```
