```python
import scipy.specialfrom pandas import *import pycosmc.post as Pimport pycosmc.derivers.hubble_theta.hubble_theta as H
```

```python
import os, cPickledef cache(name,f=None,reset=False):    try:        if reset: raise Exception()        with open('.cache/%s'%name) as f: return cPickle.load(f)    except:         assert f!=None, "Value not cached, must provide function."        r = f()        if not os.path.exists('.cache'): os.makedirs('.cache')        with open('.cache/%s'%name,'w') as f: cPickle.dump(r,f,protocol=2)        return r
```

```python
chain=P.load_chain('chains/wmapspt_tailmonty.chain')
```

```python
ps=['tau', 'theta', 'omnuh2', 'Nnu_massive', 'ombh2', 'omch2', 'logA', 'ns', 'Yp', 'nrun']
```

```python
covmat = chain.cov(ps)
```

```python
def fac(d):    """ Ratio of the volume of unit hypercube to unit hypersphere in d dimensions"""     return 2**d/(2*pi**(d/2.)/scipy.special.gamma(d/2.)/d)
```

```python
def getwhite(dim,samp,sigma):    return array(filter(lambda x: norm(x)<sigma, np.random.uniform(-sigma,sigma,(samp*fac(dim),dim))))[:samp]
```

```python
w_cps = cache('w_cps_40000_7',lambda: getwhite(dim=10,samp=40000,sigma=7))
```

```python
cps = DataMatrix(dot(w_cps,cholesky(covmat).T),columns=ps)+chain.mean(ps)
```

```python
cps=cps[cps['Yp']>.05]cps=cps[cps['Nnu_massive']>0]cps=cps[cps['omnuh2']>0]
```

```python
figsize(15,15)P.likegrid([P.Chain({k:cps[k] for k in ps}),chain],ps=ps,);
```

```python
def camb_derived(p):    h = H.theta2hubble(p["theta"],p["ombh2"],p["omch2"],0,p['omnuh2'],-1,0,p['Nnu_massive'])/100.    return {"hubble":100*h,            "ombh2":p["ombh2"],            "omch2":p["omch2"],            "omnuh2":p["omnuh2"],            "omk":0,            "scalar_amp(1)":exp(p["logA"])*10**(-10),            "scalar_spectral_index(1)":p["ns"],            "scalar_nrun(1)":p["nrun"],            "re_optical_depth":p["tau"],            "theta":p["theta"],            "logA":p["logA"],            "helium_fraction":p['Yp'],            "massless_neutrinos":0,            "massive_neutrinos":p['Nnu_massive']}
```

```python
d_cps = cache('d_cps_40000_7',lambda: DataMatrix([camb_derived(c) for _,c in cps.iterrows()]))
```

```python
def savecps(file,names,dat):    with open(file,'w') as f:        f.write('#'+' '.join(names)+'\n')        savetxt(f,[[d[n] for n in names] for _,d in dat.iterrows()])
```

```python
savecps('wmapspt_tailmonty_inner7.dat',d_cps.columns,d_cps)
```
